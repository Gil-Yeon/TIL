```python
# Use json_normalize to flatten the 'status' column
flattened_status = pd.json_normalize(ranker_df['status'])

# Concatenate the flattened status DataFrame with the original DataFrame
ranker_df = pd.concat([ranker_df, flattened_status], axis=1)

# Drop the original 'status' column if needed
ranker_df = ranker_df.drop('status', axis=1)

ranker_df
```
