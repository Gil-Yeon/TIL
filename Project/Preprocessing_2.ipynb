```python
import pandas as pd
import numpy as np
```

```python
path = './data/'

df = pd.read_csv(path+'월별데이터 총병합.csv')
df
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
school = pd.read_csv(path+'School.csv')
school
```

```python
subway = pd.read_csv(path+'Subway_Stations.csv')
subway
```

```python
univ = pd.read_csv(path+'University.csv', encoding='cp949')
univ
```

```python
park = pd.read_csv(path+'Parks.csv', encoding='cp949')
park
```

## 전처리

```python
school = school.groupby('Region_Name')['학교명'].count().reset_index()
school.rename(columns={'학교명': 'School_Counts'}, inplace=True)
school
```

```python
subway = subway.groupby('Region_Name')['역명'].count().reset_index()
subway.rename(columns={'역명': 'Subway_Counts'}, inplace=True)
subway
```

```python
univ = univ.groupby('Region_Name')['학교명'].count().reset_index()
univ.rename(columns={'학교명': 'Univ_Counts'}, inplace=True)
univ
```

```python
park = park.groupby('Region_Name')['공원명'].count().reset_index()
park.rename(columns={'공원명': 'Park_Counts'}, inplace=True)
park
```

## 병합

```python
merged_df = df

dataframes_to_merge = [school, subway, univ, park]

# 리스트에 있는 각 데이터프레임을 순차적으로 병합합니다.
for i in dataframes_to_merge:
    merged_df = pd.merge(merged_df, i, on='Region_Name', how='inner')
```

```python
merged_df.info()
```

```python
merged_df.to_csv('월별데이터 총병합.csv')
```
