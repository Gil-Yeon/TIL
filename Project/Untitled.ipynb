```python
path = './data/'
df_pop2 = pd.read_csv(path+'시군구_성_연령_5세_별_주민등록연앙인구_20230906153512.csv', encoding='EUC-KR')
```

```python
# 나이 그룹을 묶고 컬럼 이름 수정
age_groups = ['0~19세', '20~39세', '40~59세', '60~79세', '80~99세']
for age_group in age_groups:
    age_range = age_group.split('~')
    start_age, end_age = int(age_range[0]), int(age_range[1])
    columns_to_group = [f'{i}~{i+4}세' for i in range(start_age, end_age+1, 5)]
    df_pop2[age_group] = df_pop2[columns_to_group].sum(axis=1)
    df_pop2.drop(columns=columns_to_group, inplace=True)
```
