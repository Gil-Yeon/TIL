### 히스토그램
- 정상 진동 데이터 : 데이터의 분포가 중앙에 집약되어 있고 폭이 비정상 데이터보다 완만한 편이다.
- 이물질 투입 데이터 : 데이터의 분포가 넓은 범위로 분포 되어있다.
- 균형 이상 데이터 : 정상 진동 데이터와 거의 유사한 진동폭을 갖고 있으나 상대적으로 0보다 낮은 음수쪽으로 더 많은 데이터가 집약 되어있다.

### 박스 차트
- 정상 진동 데이터 : 박스의 중심이 0에 가깝게 배치되어있다.
- 이물질 투입 데이터 : 박스의 중심이 0보다 높게 위치하며, 박스의 크기가 매우 크므로 데이터가 넓은 범위로 분포되어있다는 걸 알 수 있다.
- 균형 이상 데이터 : 박스의 중심이 0보다 아래에 배치되어 있고, 정상 진동 데이터와 박스 크기가 상대적으로 작다.

### 총정리
- 정상 진동 데이터의 경우 상대적으로 중심선이 0에 가까웠고, 데이터의 집약도가 균형 이상 데이터보다는 넓은 범위로 분포되어있었다.

- 이물질 데이터의 경우 상대적으로 넓은 범위에 데이터 분포가 되있었고, 중심선은 0보다 더 위에 있는 편이다.


- 균형 이상 데이터는 다른 데이터 집단보다 데이터 집약도가 높았으며 중심선이 0보다 아래에 있는 편이다.

진동이 0에 가까울수록 기기가 정상적으로 작동된다는 것을 알 수 있었다.

```python
import numpy as np
from scipy.stats import shapiro

# A그룹 정규성 검정

# 예시 데이터
NL_A = np.array(df_A_Group['NL_A'])  # 여기에 NL_B 데이터를 입력
FB_A = np.array(df_A_Group['FB_A'])  # 여기에 FB_B 데이터를 입력
IB_A = np.array(df_A_Group['IB_A'])  # 여기에 IB_B 데이터를 입력

# Shapiro-Wilk 정규성 검정
A_stat_NL, p_NL = shapiro(NL_A)
A_stat_FB, p_FB = shapiro(FB_A)
A_stat_IB, p_IB = shapiro(IB_A)

print("NL_B Shapiro-Wilk Test: Statistic =", A_stat_NL, "P-value =", p_NL)
print("FB_B Shapiro-Wilk Test: Statistic =", A_stat_FB, "P-value =", p_FB)
print("IB_B Shapiro-Wilk Test: Statistic =", A_stat_IB, "P-value =", p_IB)

# P-value를 기반으로 정규성 판단
alpha = 0.05
if p_NL > alpha:
    print("NL_A appears to be normally distributed")
else:
    print("NL_A does not appear to be normally distributed")

if p_FB > alpha:
    print("FB_A appears to be normally distributed")
else:
    print("FB_A does not appear to be normally distributed")

if p_IB > alpha:
    print("IB_A appears to be normally distributed")
else:
    print("IB_A does not appear to be normally distributed")
```

```python
# B그룹 정규성 검정

# 예시 데이터
NL_B = np.array(df_B_Group['NL_B'])  # 여기에 NL_B 데이터를 입력
FB_B = np.array(df_B_Group['FB_B'])  # 여기에 FB_B 데이터를 입력
IB_B = np.array(df_B_Group['IB_B'])  # 여기에 IB_B 데이터를 입력

# Shapiro-Wilk 정규성 검정
B_stat_NL, p_NL = shapiro(NL_B)
B_stat_FB, p_FB = shapiro(FB_B)
B_stat_IB, p_IB = shapiro(IB_B)

print("NL_B Shapiro-Wilk Test: Statistic =", B_stat_NL, "P-value =", p_NL)
print("FB_B Shapiro-Wilk Test: Statistic =", B_stat_FB, "P-value =", p_FB)
print("IB_B Shapiro-Wilk Test: Statistic =", B_stat_IB, "P-value =", p_IB)

# P-value를 기반으로 정규성 판단
alpha = 0.05
if p_NL > alpha:
    print("NL_B appears to be normally distributed")
else:
    print("NL_B does not appear to be normally distributed")

if p_FB > alpha:
    print("FB_B appears to be normally distributed")
else:
    print("FB_B does not appear to be normally distributed")

if p_IB > alpha:
    print("IB_B appears to be normally distributed")
else:
    print("IB_B does not appear to be normally distributed")
```

```python
# scipy.stats 라이브러리에서 shapiro 함수(Shapiro-Wilk 정규성 검정 함수)를 가져옴
from scipy.stats import shapiro

# 유의 수준을 0.05로 설정
alpha = 0.05

# df의 모든 컬럼(변수)에 대해 반복문을 사용하여 정규성 검정을 수행
for column in df.columns:
    # 각 컬럼에 대해 Shapiro-Wilk 정규성 검정을 실행
    stat, p = shapiro(df[column])

    # 검정 결과를 출력
    print(f"{column} Shapiro-Wilk Test: Statistic = {stat}, P-value = {p}")

    # P-value가 alpha보다 크면 데이터가 정규 분포를 따른다고 판단
    if p > alpha:
        print(f"{column} appears to be normally distributed")
    # 그렇지 않으면 정규 분포를 따르지 않는다고 판단
    else:
        print(f"{column} does not appear to be normally distributed")

    # 결과를 구분하기 위해 '-----' 출력
    print('-' * 50)
```

```python
# scipy.stats 라이브러리에서 f_oneway 함수를 불러옵니다.
# 이 함수는 일원 분산분석(ANOVA)을 실행하기 위한 함수입니다.
# H0 : 그룹 간 평균치 차이가 없다.
# H1 : 그룹 간 평균치 차이가 있다.
from scipy.stats import f_oneway

# 데이터프레임 df에서 'NL_A', 'FB_A', 'IB_A' 컬럼의 데이터를 각각의 변수에 저장합니다.
NL_A_data = df['NL_A']
FB_A_data = df['FB_A']
IB_A_data = df['IB_A']

# 일원 분산분석을 실행하여, F-통계량 값(f_stat)과 p-value(p_val)를 얻습니다.
f_stat, p_val = f_oneway(NL_A_data, FB_A_data, IB_A_data)

# 결과를 출력합니다.
print("Results for comparing NL_A, FB_A, and IB_A:")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}\n")

# F-statistic: 0.0748
#   F-통계량은 그룹 간의 평균값 차이와 그룹 내에서의 분산을 비교한 값 입니다.
#   높은 F-통계량 값은 그룹간의 평균값이 서로 다르다는 것을 나타내며, 낮은 값은 평균값이 비슷하다는 것을 의미합니다.

# P-value: 0.9280 = 귀무가설을 채택하므로 그룹간의 평균값 차이가 통계적으로 유의미하지 않다고 볼 수 있다.
#   P-value는 관찰된 데이터나 더 극단적인 데이터가 주어진 가설 하에서 발생할 확률을 나타냅니다.
#   일반적으로 0.05(5%)보다 작으면, 그룹간의 평균값 차이가 유의미하다고 판단합니다.
```
