# 데이터 불러오기

```python
df = pd.read_csv('./data/서울시5대범죄.csv', encoding='UTF-8')
df.head()
```

```python
df.info()
```

# 범죄율, 범죄율 순위 구하기
- 인구대비범죄율 (범죄 발생 건수 / 인구수) * 10만
- 범죄 발생 건수의 차이를 비교하기 쉽기 때문에 10만명으로 잡는다

```python
# 범죄율 구하기(소수점 2자리로)
df['인구대비범죄율'] = (df['발생합계'] / df['인구수'] * 100000).round(2)

print(df) 
```

```python
# 범죄율순위 구하기
df["범죄율등수"] = (df.groupby("년도")['인구대비범죄율'].rank(ascending=False).astype(int))

df
```

# 수정된 데이터프레임 저장하기(.CSV)

```python
df.to_csv('서울시5대범죄전처리.csv')
```
