## 문제 1.
- 주어진 trash bag 데이터 세트는 지역별 종량제 봉투 가격을 나타낸다. 가격 컬럼은 각 행의 조건을 만족하는 해당 용량의 종량제 봉투가 존재하면 가격을 값으로, 존재하지 않으면 0을 값으로 갖는다. 이 때 용도가 '음식물쓰레기'이고 사용 대상이 '가정용'인 2L 봉투 가격의 평균을 소수점을 버린 후 정수로 출력한다. 

```python
import pandas as pd 
df = pd.read_csv("data/5회/501_trash_bag.csv", encoding="euc-kr")
df.head()
```

```python
df['용도'].value_counts()
```

```python
df['사용대상'].value_counts()
```

```python
data = df[(df['용도']=='음식물쓰레기')&(df['사용대상']=='가정용')]['2L가격'].reset_index(drop=True)
df1 = pd.DataFrame(data)
df1
```

```python
df2 = df1[df1['2L가격'] != 0]
result = int(df2.mean())
print(result)
```

## 문제 2. 
- BMI지수는 몸무게(kg)를 키(m)의 제곱으로 나누어 구하며, BMI값에 따른 비만도 분류는 다음과 같다.

```python
import pandas as pd
pd.DataFrame({
    "BMIT지수 범위" : ["18.5 미만", "18.5 이상 23미만", "23 이상 25 미만", "25이상 30미만", "30이상"], 
    "비만도 분류" : ["저체중", "정상", "과체중", "경도비만", "중등도비만"]
})
```

- 이 때 주어진 BMI 데이터 세트에서 비만도가 정상에 속하는 인원 수와 과체중에 속하는 인원 수의 차이를 정수로 출력하시오. 

```python
import pandas as pd 
df = pd.read_csv("data/5회/502_bmi.csv", encoding="euc-kr")
df
```

```python
df['BMI'] = df['Weight']/(df['Height']/100)**2
df
```

```python
normal = df[(df['BMI']>=23) & (df['BMI']<25)]
over = df[(df['BMI']>=18.5) & (df['BMI']<23)]
```

```python
result = abs(len(normal)-len(over))
print(result)
```

## 문제 3. 
- 주어진 students 데이터 세트는 각 학교의 학년별 총 전입학생, 총 전출학생, 전체 학생 수를 나타낸다. 순 전입학생수는 총 전입학생 수에서 총 전출학생 수를 빼서 구할 수 있다. 순 전입학생이 가장 많은 학교의 전체 학생 수를 구하시오. 

```python
import pandas as pd 
df = pd.read_csv("data/5회/503_students.csv", encoding="euc-kr")
df['순 전입학생'] = df['총 전입학생'] - df['총 전출학생']
df
```

```python
rst_df = df.groupby('학교').agg(sum).sort_values('순 전입학생', ascending=False).reset_index()
rst_df
```

```python
result = rst_df.iloc[0]['전체 학생 수']
result
```
