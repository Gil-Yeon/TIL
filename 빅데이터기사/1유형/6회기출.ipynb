## Q1. 주어진 데이터를 이용하여 이동시간(출동시간과 도착시간의 차이)이 평균적으로 가장 오래 걸린 구급대의 이동시간을 구하시오. (분 단위로 환산 및 반올림하여 정수로 출력)
- 데이터 : data_6_1_1.csv

### 풀이1

```python
import pandas as pd
df = pd.read_csv('data/6회/data_6_1_1.csv')
df.head()
```

```python
df['출동시간'] = pd.to_datetime(df['출동시간'])
df['도착시간'] = pd.to_datetime(df['도착시간'])
df['이동시간'] = df['도착시간'] - df['출동시간']
df.head()
```

```python
rst_df = df.groupby('구급대')['이동시간'].agg('mean').reset_index().sort_values('이동시간', ascending=False)
rst_df
```

```python
result = round(rst_df.iloc[0]['이동시간'].seconds/60)
print(result)
```

```python
rst_df['이동시간'].dt.total_seconds()
```

### 풀이2

```python
df['출동시간'] = pd.to_datetime(df['출동시간'])
df['도착시간'] = pd.to_datetime(df['도착시간'])
df['이동시간'] = df['도착시간'] - df['출동시간']
df.head()
```

```python
rst_df = df.groupby('구급대')['이동시간'].agg('mean').reset_index().sort_values('이동시간', ascending=False)
rst_df
```

```python
rst_df['이동시간2'] = rst_df['이동시간'].dt.seconds
rst_df
```

```python
result = round(rst_df.iloc[0]['이동시간2']/60)
print(result)
```

## Q2. 주어진 데이터를 이용하여 교수 1인당 맡고 있는 학생 수가 가장 많은 대학교를 찾고, 그 대학교의 전체 교수의 수를 구하시오. (정수로 출력)
- 데이터 : data_6_1_2.csv

### 풀이1

```python
import pandas as pd
df = pd.read_csv('data/6회/data_6_1_2.csv')
df.head()
```

```python
df['1인당 학생수'] = (df['경영학과']+df['경제학과']+df['컴퓨터공학과']+df['산업공학과']+df['국문학과']+df['영문학과'])/df['교수']
df.head()
```

```python
df_sorted = df.sort_values(by='1인당 학생수', ascending=False)
df_sorted.head()
```

```python
result = df_sorted.iloc[0]['교수']
print(result)
```

### 풀이2

```python
import pandas as pd
df = pd.read_csv('data/6회/data_6_1_2.csv')
df.head()
```

```python
df['1인당 학생수'] = df.iloc[:, 2:].sum(axis=1)/df['교수']
df.head()
```

```python
df_sorted = df.sort_values(by='1인당 학생수', ascending=False)
df_sorted.head()
```

```python
result = df_sorted.iloc[0]['교수']
print(result)
```

## Q3. 주어진 데이터를 이용하여 연도별 총 범죄 건수의 월평균 값을 구하고, 그 값이 가장 큰 연도의 총 범죄 건수의 월평균 값을 집계하시오. (반올림하여 정수로 출력)
- 데이터 : data_6_1_3.csv

### 풀이1

```python
import pandas as pd

df = pd.read_csv('data/6회/data_6_1_3.csv')
df.head()
```

```python
df['총범죄'] = df.iloc[:, 1:-1].sum(axis=1)
df.head()
```

```python
df['year'] = df['날짜'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
result_df = df.groupby('year')['총범죄'].mean()
result_df
```

```python
result = round(result_df.max())
print(result)
```
