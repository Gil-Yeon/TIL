# 1. 빅데이터 시대의 현재와 미래

## 빅데이터란

# 2. 분석, 빅데이터 시대 필수역량

## 문제인식의 6단계
- 문제인식 : 개인이 관심있거나 하고있는 분야에서 문제를 인식하고 그것을 해결하려는 것에서 시작
- 관련 연구 조사 : 문제와 직접적, 간접적으로 관련된 지식을 논문이나 책 등에서 찾아 문제와 관련된 주요 변수들을 파악하는 단계
- 모형화 : 문제와 관련된 내용을 정리해 문제 해결에 직접적인 관련이 있는 변수를 뽑아내는 단계
- 자료수집 : 모형화를 통해 주요 변수로 재구성되고 측정이라는 과정을 거쳐 자료를 수집하는 단계
- 자료분석 : 나열된 숫자에서 변수 사이의 규칙적인 패턴인 변수 간 관련성을 파악하는 단계
- 결과제시 : 다양한 차트나 그래프를 통해 결과를 제시하는 단계

# 4. 퍼센트를 조심하라

## 퍼센트
- 무엇에 대한 퍼센트라고 표현할 때 그 무엇이 언제나 기준이 되며, 이 기준은 퍼센트에서 분모가 된다.
- 퍼센트는 2개 이상의 숫자의 상대적 크기를 명확하게 하기 위해 사용된다.


## 퍼센트 유의사항
- 퍼센트를 대할 때는 무엇에 대한 포인트이고, 기준이 제대로 적용되었는지 항상 따져봐야 한다.
- 어떻게 기준을 잡느냐에 따라 눈속임이 가능하므로 주의한다.

## 퍼센트 포인트
- 두 퍼센트의 기준이 같아 보통의 숫자와 같이 비교가 가능할 때 %p라고 표기하고 퍼센트 포인트라 읽는다.

# 5. 시각화의 오용과 남용


## 시각화의 중요성
- 일반적으로 그래프를 많이 사용함
- 숫자가 나타내는 정보를 시각화하여 이해하기 쉽도록 전달해야 함

## 그래프의 함정
- 그래프를 그리는 사람의 의도가 반영될 수 있으므로, 그래프 눈금의 크기를 일관성 있게 유지해야 한다.
- 그래프를 제대로 보기 위해서는 그래프가 전체 그림인지, 눈금이 과장되어 있진 않은지를 고려해야 한다.
- 그림도표의 경우 그리는 과정에서의 과장이나 축소에 주의해야 한다.

# 6. 확률에도 종류가 있다


## 선험적 확률
- 경험하지 않고도 이론적으로 미리 알 수 있는 확률
  - ex) 동전의 앞 뒷면

## 경험적 확률
- 경험적 확률 : 오랜 기간에 걸쳐 동일한 상황이나 조건하에서 어떤 사건이 일어나는 상대적인 비율로서 확률을 해석하는 것
  - ex) 벤포드 법칙 -> 데이터 조작탐지에 사용

## 주관적 확률
- 개인이 어떤 사건이 일어날 것이라고 기대하는 확률로 주관적 확률의 객관성, 정확성이 의사결정의 성공여부를 좌우한다.

# 7. 확률을 이용한 판단 오류

## 도박사의 오류
- 시도를 반복할수록 이론적인 확률에 접근한다는 평균의 법칙 또는 대수의 법칙으로 인해 독립적인 사건들을 종속적인 사건으로 바라보고 다음번에는 어떤 결과가 나올 확률이 높다고 여기는 것


## 유용성의 오류
- 쉽게 기억나는 사건들이 일어날 확률을 높게 평가하는 것
  - ex) 머피의 법칙

# 8. 여론조사 올바로 이해하기

## 표본조사와 대표성
- 모집단에서 뽑은 표본은 모집단의 대표성을 가지고 있어야 한다.

## 표본의 무작위 추출
- 모집단에 속한 대상들이 표본에 뽑힐 확률이 모두 동일한 것
- 이 방식으로 표본을 뽑을 때는 사람의 판단이나 편리함이 전혀 고려되지 않는다.
- 편의표본 : 표본을 뽑는 사람의 판단이나 편리함이 고려된 표본으로, 편중될 수 있으므로 유의해야 한다.
- 표본크기 결정 요인 : 모집단이 얼마나 다양한지, 조사가 어느 정도 정확도를 요구하는지에 따라 다르다.

## 질문과 응답항목
- 질문에 어떤 가치 판단을 깔아놓고 답을 원하는 방향으로 유도할 수있으므로, 질 좋은 질문을 해야한다.
- 특정한 응답에 대한 응답비율을 높이려는 조사자의 의도가 배제되어야 한다.

## 조사결과의 해석
- 확대해석으로 조사결과가 달라질 수 있으므로 이를 유의하여 살펴보아야 한다.
- 모집단의 정의, 표집방법, 응답률, 표본의 크기를 체크하여 표본의 대표성이 유지되어 표본이 적절한지 체크해야 한다.

# 9. 상관관계는 인과관계가 아니다

## 인과관계
- 원인은 결과보다 시간적으로 앞서야 한다.
- 원인과 결과는 관련이 있어야 한다.
- 결과는 원인이 되는 변수만으로 설명되어야 하고, 다른 변수에 의한 설명은 제거되어야 한다.

## 상관관계
- 어떤 변수가 증가할 때 다른 변수도 함께 증가하거나 감소하는 관계
- 두 변수 간 관계가 밀접하다는 것을 나타낼 뿐, 인과관계의 증거가 되진 않는다.
- 단순 상관관계를 인과관계로 해석하지 않도록 주의
  - ex) 아이스크림 판매량, 상어에 의한 피해

## 우연한 상관관계
- 허위관련성 : 두 변수를 조사했을 때 두 변수가 그냥 우연히 상관관계를 나타내는 경우가 흔하다는 것
- 전후인과의 오류 : A가 일어난 다음 B가 일어났다고 해서 A가 B의 원인이라고 결론 짓는 것

## 인과관계의 방향
- 상관관계가 있다고 할 때 흔히 생각하는 것과 반대로 인과관계가 작용할 수도 있다.
- 흡연이 폐암발생의 유일한 원인인 것처럼 성급하게 단순화하지 않도록 유의한다.

# 10. 비교, 어떻게 해야 할까?

## 비교의 함정
- 비교란 둘 이상의 사물을 견주어 서로 간의 유사점, 차이점, 일반 법칙 따위를 고찰하는 것이다.
- 비교 대상을 계량화하기 어려운 경우 주관적인 판단 하에 비교가 이루어지기 때문에 정확한 비교가 어렵다.

## 올바로 비교하는 법
- 비교되는 특성에 대한 정의가 같아야 한다.
  - ex) 우리나라의 실업률과 미국의 실업률을 비교할 때 각 나라의 실업률에 대한 정의가 다르기 때문에 옳지 않다.
- 비교되는 특성 이외의 것들에 대해 서로 비슷해야 한다.
  - ex) 미 해군 사망률과 뉴욕시 사망률을 따지는 경우 비교되는 특성인 사망률 외에 미 해군의 구성인원(청년들)과 뉴욕시의 구성인원(노약자 포함)이 서로 다르므로 옳지 않다.

# 11. 어림수의 허와 실

## 어림수
- 숫자는 과학적이라는 이미지 때문에 어림수를 사용하여 이야기하는 경우가 많다.
- 여러 가지 한계 때문에 정확한 숫자의 계산이 어려운 경우 어림수를 사용한다.

어림수의 허
- 대응법
  - 상식적으로 생각한다.
  - 어림수의 근거에 대해 질문한다.
  - 상대방의 주장을 입증하는 추가 증거가 있는지 확인한다.
- 모든 숫자를 의심하는 과정 -> 이 과정으로 확신을 얻음
  - 관련성 : 숫자가 의미를 가지려면 해당 주제와 직접적으로 관련되어야 한다.
  - 정확성 : 누가 어떻게 그 숫자를 만들어냈고, 왜 그런 방법을 사용했는지, 어떤 의도가 개입된 것인지 생각함으로써 숫자의 정확성이 판단 가능하다.
  - 올바른 해석 : 숫자의 해석이 해결하려는 문제와 관련지어 적절한지를 의심한다.

# 12. 창의성과 분석

## 창의성은 분석에서 온다.
- 예술 영역을 제외한 대부분의 영역에서 계량적 분석이 창의성의 원천이다.
- 분석을 성공적으로 수행하는 것은 결과적으로는 창의력을 발휘하는 것이다.
- 분석 측면에서 '문제 해결과 관련된 변수를 선정하여 측정한 뒤 변수들 간의 새로운 관련성을 파악하는 능력'으로 창의성을 정의할 수 있다.

## 분석능력 키우기
- 관련 연구를 모두 조사하고, 선정된 변수를 측정하여 변수 간의 새로운 관련성을 파악해야 분석능력을 키울 수 있다.
- 오랜 경험을 통해 뛰어난 분석능력을 키우면 창의성, 영감, 직관 등의 능력을 가질 수 있다(신경세포는 계속적인 분석에 의해서 훈련될 수 있기 때문이다)

## 창의성
- 창의성이란 우리가 아직 파악하지 못한 변수 간의 규칙적인 패턴을 찾아내 이를 유용하게 활용하는 것이다.
