# 빅데이터 시대의 현재와 미래

## 빅데이터란
- 기존의 IT 기술로는 처리하기가 어려운 데이터
- 빅데이터 활용을 위해서는 고객과 시장에 대한 Insight도출이 필요
- 적절한 기계학습으로 자동적으로 Insight를 도출하여 의사결정에 활용한다.

## 빅데이터를 통한 경쟁우위
- 데이터에 바탕을 둔 사실에 근거한 의사결정이 필요하다.
- 데이터 분석으로 고객과 시장에 대한 Insight를 도출하고 현명한 의사결정을 해야 지속가능한 경쟁우위 창출이 가능하다.

## 빅데이터 시대 대응방안
- 기업의 경영진은 분석지향 리더십이 필요
- 사업 전반을 디지털로 혁신하는 것이 필요

# 분석, 빅데이터 시대 필수역량

## 문제인식의 6단계
- 문제인식 : 개인이 관심있거나 하고있는 분야에서 문제를 인식하고 그것을 해결하려는 것에서 시작
- 관련 연구 조사 : 문제와 직접적, 간접적으로 관련된 지식을 논문이나 책 등에서 찾아 문제와 관련된 주요 변수들을 파악하는 단계
- 모형화 : 문제와 관련된 내용을 정리해 문제 해결에 직접적인 관련이 있는 변수를 뽑아내는 단계
- 자료수집 : 모형화를 통해 주요 변수로 재구성되고 측정이라는 과정을 거쳐 자료를 수집하는 단계
- 자료분석 : 나열된 숫자에서 변수 사이의 규칙적인 패턴인 변수 간 관련성을 파악하는 단계
- 결과제시 : 다양한 차트나 그래프를 통해 결과를 제시하는 단계

# 퍼센트를 조심하라

## 퍼센트
- 무엇에 대한 퍼센트라고 표현할 때 그 무엇이 언제나 기준이 되며, 이 기준은 퍼센트에서 분모가 된다.
- 퍼센트는 2개 이상의 숫자의 상대적 크기를 명확하게 하기 위해 사용된다.


## 퍼센트 유의사항
- 퍼센트를 대할 때는 무엇에 대한 포인트이고, 기준이 제대로 적용되었는지 항상 따져봐야 한다.
- 어떻게 기준을 잡느냐에 따라 눈속임이 가능하므로 주의한다.

## 퍼센트 포인트
- 두 퍼센트의 기준이 같아 보통의 숫자와 같이 비교가 가능할 때 %p라고 표기하고 퍼센트 포인트라 읽는다.

# 시각화의 오용과 남용


## 시각화의 중요성
- 일반적으로 그래프를 많이 사용함
- 숫자가 나타내는 정보를 시각화하여 이해하기 쉽도록 전달해야 함

## 그래프의 함정
- 그래프를 그리는 사람의 의도가 반영될 수 있으므로, 그래프 눈금의 크기를 일관성 있게 유지해야 한다.
- 그래프를 제대로 보기 위해서는 그래프가 전체 그림인지, 눈금이 과장되어 있진 않은지를 고려해야 한다.
- 그림도표의 경우 그리는 과정에서의 과장이나 축소에 주의해야 한다.

# 확률에도 종류가 있다


## 선험적 확률
- 경험하지 않고도 이론적으로 미리 알 수 있는 확률
  - ex) 동전의 앞 뒷면

## 경험적 확률
- 경험적 확률 : 오랜 기간에 걸쳐 동일한 상황이나 조건하에서 어떤 사건이 일어나는 상대적인 비율로서 확률을 해석하는 것
  - ex) 벤포드 법칙 -> 데이터 조작탐지에 사용

## 주관적 확률
- 개인이 어떤 사건이 일어날 것이라고 기대하는 확률로 주관적 확률의 객관성, 정확성이 의사결정의 성공여부를 좌우한다.
